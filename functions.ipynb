# Functions

Python Functions are a block of statements that does a specific task. The idea is to put some commonly or repeatedly done task together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.

Lets create a function(with docstring)


syntax :

def functionname (parameter):

        " docstring for
        fucntion info"

        body of function

        return function

functionName(input) # callin the function

```python
def is_even(num):
  """
  This function returns if a given number is odd or even
  input - any valid integer
  output - odd/even
  """
  if num % 2 == 0:
    return 'even'
  else:
    return 'odd'
```

```python
# function
# function_name(input)

for i in range(1,11):
  x = is_even(i)
  print(x)
```

*agrs and **kwargs

*args and **kwargs are special python keywords that are used to pass the variable length of arguments to a function

```python
# *args
# allows us to pass a variable number of non-keyword argument to a function

def multiply(*args):
  product = 1
  for i in args:
    product *= i

  return product
```

```python
multiply(2,3,4)
```

```python
# **kwargs
# allows us to pass any number of keyword arguments.
# keywords arguments means that they contain a key-value pair like a Python dictionary.

def display(**kwargs):

  for (key,value) in kwargs.items():
    print(key,'->',value)
```

```python
display(India = 'Delhi', SriLanka = 'colambo')
```

# Points to remember while using *args and **kwargs

  * Order of the arguments matters(normal -> *args -> **kwargs)

  * The words "args" and "kwargs" are only a convention, you can use any name of your choice.

```python
def g(y):
  print(x)
  print(x+1)

x = 5
g(x)
print(x)
```

```python
def f(y):
  x = 1
  x += 1
  print(x)

x = 5
f(x)
print(x)
```

```python
def h(y):
  x += 1
x = 5
h(x)
print(x)

# we can not change global variable from local function
```

```python
def h(y):
  global x
  x += 1

x = 5
h(x)
print(x)
```

```python
def f(x):
  x = x + 1
  print('in f(x): x =', x)
  return x

x = 3
z = f(x)
print('in main program scope:', z)
print('in main program scope:', x)
```

# Lambda function

```python
a = lambda x : x**2
a(4)
```

```python
a = lambda x,y : x+y
a(4,5)
```

```python
a = lambda s: 'a' in s
a('hay')
```

# Higher order function

```python
def square(x):
  return x**2

def transform(f,l):
  output = []
  for i in l:
    output.append(f(i))

  print(output)

l= [1,2,3,4,5]
transform(square,l)
```

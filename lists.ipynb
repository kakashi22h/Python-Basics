#Creating a List

```python
#Empty
print([])
# 1D
print([1,2,3,4])
# 2D
print([1,3,4,[2,5]])
# 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
# Hetrogenous
print([1,2,'true',True,4.5])
# Using Type conversion
print(list('hello'))
```

#Accessing Items from a list

```python
# Indexing
l = [1,2,3,[4,5]]
m = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(m[0][0][1])
print(l[0])
print(l[3][1])

# Slicing
L = [1,2,3,4,5,6,7,7,8,8,9]
print(L[0:3])
print(L[-3:])
print(L[-5:2:2])
print(L[::-1])
```

#Adding Items to a List

```python
# append
L = [1,2,3,4,5]
L.append(6)
print(L)
```

```python
# extend
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
# insert
L = [1,2,3,4,5]
L.insert(1,100)
print(L)
```

#Editing items in a List

```python
L = [1,2,3,4,5]

L[-1]= 500
print(L)
L[1:4] = [200,300,400]
print(L)
```

#Deleting item from a List

```python
# del
L = [1,2,3,4]
print(L)
# del L[-1]
# del L[1:3]
del L
print(L)
```

```python
# remove
L = [1,2,3,4,5]
L.remove(5)
print(L)
```

```python
# pop
L = [1,2,3,4,5]
L.pop()
print(L)
L.pop(2)
print(L)
```

```python
# clear
L = [1,2,3,4,5]
print(L)
L.clear()
print(L)
```

#Operations on List

  * Arithmetic
  * Membership
  * Loop

```python
# Arithmetic (+.*)
L1 = [1,2,3,4]
L2 = [5,6,7,8]

print(L1 + L2)
```

```python
print(L1*3)
```

```python
# Membership
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,5,6]

print(5 in L1)
print(6 not in L2)
```

```python
# Loops
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,5,6]

for i in L1:
  print(i)
```

#List functions

```python
# len/min/max/sorted
L = [2,1,5,7,0]
print(len(L))
print(min(L))
print(max(L))
print(sorted(L))
print(sorted(L,reverse=True))
```

```python
# count
L = [1,2,2,3,4,2,5]
L.count(2)
```

```python
# index
L = [1,2,3,4,5]
L.index(2)
```

```python
# reverse
l = [1,2,3,4]
# it permanently reverse a list
l.reverse()
```

```python
# sort ( vs sorted)
l = [5,6,7,1,2,3]
# sort is permanent
l.sort()
print(l)
```

```python
# copy
l = [1,2,3,4]
print(id(l))
L1 = l.copy()
print(id(L1))
```

#List Comprehension

List Comprehension provides a concise way of creating lists.
newlist = [expression for item in iterable if condition == True]

```python
# Add 1 to 10 numbers to a list
L = []

for i in range(1,11):
  L.append(i)
print(L)
```

```python
L = [i for i in range(1,11)]
print(L)
```

```python
v = [2,3,4]
s = -3

L = [s*i for i in v ]
print(L)
```

```python
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
[i for i in range(1,51) if i % 5 == 0]
```

```python
languages = ['java','python','php','c','js']
[language for language in languages if language.startswith('p')]
```

```python
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
[[i*j for i in range(1,4)] for j in range(1,4)]
```
